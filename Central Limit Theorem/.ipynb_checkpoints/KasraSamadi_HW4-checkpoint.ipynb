#### تئوری Central Limit را روی توزیعهای مختلف داده شامل چوله به چپ، راست، یونیفورم و نرمال بررسی کنید. برای این کار ابتدا به ازای هر یک از شکل های توزیع ذکر شده لیستی از صد هزار عدد را به وسیله زبان پایتون ایجاد کنید. سپس به ازای تغییر دو فاکتور مهم در این تئوری شامل تعداد سمپل ها (k) و تعداد نمونه های هر سمپل (n) بررسی کنید که نمودار توزیع میانگین این سمپل ها چه شکلی می گیرد. حاصل بررسی خود را در قالب یک فایل ورد شامل هیستوگرام توزیع مجموعه داده ایجاد شده، به علاوه هیستوگرام های حاصل از تغییرات اعمال شده روی n و k و توضیحات مکفی به ازای هر کدام و در نهایت شرحی از نتیجه گیریتان از این تئوری در پاسخ به این تمرین ارسال نمایید.

## Import Libraries

```python
import matplotlib.pyplot as plt
import numpy as np
from scipy.stats import  uniform, norm, expon, lognorm
from scipy.stats import gamma
```

```python
nums = 100000
```

```python
normal_data = norm.rvs(size=nums) # نرمال
uniform_data = uniform.rvs(size=nums) # یونیفرم
right_skewed_data = 1 / gamma.rvs(2, scale=2, size=nums) # چوله به راست
left_skewed_data = expon.rvs(scale=1, size=nums) # چوله به چپ
```

```python
num_samples = [10, 20, 30, 35, 40, 50] # k تعداد سمپل ها     
sample_sizes = [7, 8, 9, 10 ,20] # n تعداد نمونه‌های هر سمپل
```

```python
plt.hist(normal_data, bins=50, density=True, alpha=0.7)
plt.title('Normal Distribution')
plt.xlabel('Values')
plt.ylabel('Probability Density')
plt.show()
```

```python
plt.hist(uniform_data, bins=50, density=True, alpha=0.7)
plt.title('Uniform Distribution')
plt.xlabel('Values')
plt.ylabel('Probability Density')
plt.show()
```

```python
plt.hist(right_skewed_data, bins=30, density=True, alpha=0.7)
plt.title('Right Skewed Distribution')
plt.xlabel('Values')
plt.ylabel('Probability Density')
plt.show()
```

```python
plt.hist(left_skewed_data, bins=50, density=True, alpha=0.7)
plt.title('Left Skewed Distribution')
plt.xlabel('Values')
plt.ylabel('Probability Density')
plt.show()
```

```python
# تابعی برای رسم نمودار توزیع میانگین سمپل‌ها
def plot_sample_mean_distribution(means, title):
    plt.hist(means, bins=50, density=True, alpha=0.7)
    plt.title(title)
    plt.xlabel('Sample Means')
    plt.ylabel('Probability Density')
    plt.show()
```

```python
for k in num_samples:
    for n in sample_sizes:
        sample_means = []
        for _ in range(k):
            sample = np.random.choice(normal_data, size=n)
            sample_means.append(np.mean(sample))
        
        title = f'sample mean normal distribution k={k} n={n}'
        plot_sample_mean_distribution(sample_means, title)
        
```

```python
for k in num_samples:
    for n in sample_sizes:
        sample_means = []
        for _ in range(k):
            sample = np.random.choice(uniform_data, size=n)
            sample_means.append(np.mean(sample))
        
        title = f'sample mean uniform distribution k={k} n={n}'
        plot_sample_mean_distribution(sample_means, title)
```

```python
for k in num_samples:
    for n in sample_sizes:
        sample_means = []
        for _ in range(k):
            sample = np.random.choice(right_skewed_data, size=n)
            sample_means.append(np.mean(sample))
        
        title = f'sample mean right_skewed distribution k={k} n={n}'
        plot_sample_mean_distribution(sample_means, title)
```

```python
for k in num_samples:
    for n in sample_sizes:
        sample_means = []
        for _ in range(k):
            sample = np.random.choice(left_skewed_data, size=n)
            sample_means.append(np.mean(sample))
        
        title = f'sample mean left_skewed distribution k={k} n={n}'
        plot_sample_mean_distribution(sample_means, title)
```
